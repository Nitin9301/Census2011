```python
# Grouping the data by state and summing the population
state_population = census_data.groupby('State name')['Population'].sum().reset_index()

# Finding the state with the highest population
highest_population_state = state_population.loc[state_population['Population'].idxmax()]

print("State with the highest population:")
print(highest_population_state)

# Plotting the population by state
plt.figure(figsize=(14, 8))
sns.barplot(x='State name', y='Population', data=state_population, palette='viridis')
plt.xticks(rotation=90)
plt.title('Population by State')
plt.xlabel('State')
plt.ylabel('Population')
plt.show()
```

```python
# Group the data by state and sum the populations
state_population = census_data.groupby('State name')['Population'].sum()

# Plotting the pie chart
plt.figure(figsize=(12, 8))
plt.pie(state_population, labels=state_population.index, autopct='%1.1f%%', startangle=140)
plt.title('Population Distribution by State')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
